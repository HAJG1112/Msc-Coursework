```python
df_train = pd.read_csv('../input/train.csv')
```

```python
df_train.columns
```

```python
survived = df_train[df_train['Survived'] == 1]
```

Now, we create a column which gives a value of 1 if they are under or equal to 39 years of age, and a value of 0 if they are not. We also note that a person survives if survived = 1

```python
df_train['tn'] = df_train['Age']<=39   # return true if they are aged 39 or less
df_train['tn'] = df_train['tn'].astype(int)   #return true = 1 and false = 0
df_train
```

```python
np.where((df_train['tn']==df_train['Survived']), 1,0).mean()
```

```python
sns.barplot(x='tn', y='Survived', data=df_train)
```
